```python
import folium
import os
```

```python
antarctic_ice_edge = os.path.join('antarctic_ice_edge.json')
antarctic_ice_shelf_topo = os.path.join( 'antarctic_ice_shelf_topo.json')
```

```python
m = folium.Map(
    location=[51.509091, -0.124038],
    tiles='Mapbox Bright',
    zoom_start=10
)

folium.GeoJson(
    antarctic_ice_edge,
    name='geojson'
).add_to(m)



folium.LayerControl().add_to(m)


m
```

```python
m = folium.Map(
    location=[-59.1759, -11.6016],
    tiles='Mapbox Bright',
    zoom_start=2
)

folium.GeoJson(
    antarctic_ice_edge,
    name='geojson'
).add_to(m)

folium.TopoJson(
    open(antarctic_ice_shelf_topo),
    'objects.antarctic_ice_shelf',
    name='topojson'
).add_to(m)

folium.LayerControl().add_to(m)


m
```
